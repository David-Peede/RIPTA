# Generating Tutorial Data
First, we load the necessary packages.

```python
# Import packages.
import demes
import demesdraw
import gzip
from matplotlib import pyplot as plt
import msprime
import numpy as np
import tskit
# Print versions of our libraries.
print('demes', demes.__version__)
print('demesdraw', demesdraw.__version__)
print('msprime', msprime.__version__)
print('tskit', tskit.__version__)
```

Next, we load the demographic model and print the demography debugger as a sanity check.

```python
# Load the yaml from stdpopsim.
papuan_yaml = demes.load('./HomSap__PapuansOutOfAfrica_10J19.yaml')
# Load as a demographic object.
papuan_demography = msprime.Demography.from_demes(papuan_yaml)
# Show the demography debugger.
papuan_demography.debug()
```

Next, let's plot the demographic model just as an added sanity check.

```python
# Intialize the figure.
fig = plt.figure(figsize=(6, 4), constrained_layout=True, dpi=300)
# Intialize the axes.
ax = fig.add_subplot(111)
# Plot the demographic model.
demesdraw.tubes(papuan_yaml, log_time=True, ax=ax)
# Show the plot!
plt.show()
```

Now that we are confident in our demographic model, it is time to simulate!

```python
# Simulate a genomic region comprable to chromosome 22.
ts = msprime.sim_ancestry(
    samples=[
        msprime.SampleSet(10, ploidy=2, population='YRI'),
        msprime.SampleSet(10, ploidy=2, population='Papuan'),
        msprime.SampleSet(1, ploidy=2, population='NeaA'),
        msprime.SampleSet(1, ploidy=2, population='DenA'),
        ],
    demography=papuan_demography,
    sequence_length=51_304_566, recombination_rate=1.4445 * 10**-8,
    random_seed=42,
)
# Overlay mutations on the tree-sequence.
mts = msprime.sim_mutations(
    tree_sequence=ts, rate=1.29 * 10**-8,
    model='jc69', random_seed=42,
)
# Display the tree-sequence summary.
mts
```

Now, we convert and dump the tree-sequence to a gzipped vcf file—__NOTE__ this is not a bgzipped vcf file so it won't work tools such as `tabix` but that is ok since the data is just for this tutorial :)

```python
# Intialize the YRI individual IDs.
yris = ['YRI_{0}'.format(i) for i in range(10)]
# Intialize the PAP individual IDs.
paps = ['PAP_{0}'.format(i) for i in range(10)]
# Intialize the archaic individual IDs.
arcs = ['NEA', 'DEN']
# Open a vcf file.
with gzip.open('./tutorial_data.vcf.gz', 'wt') as vcf_file:
    # Write the contents of the tree-sequence to the vcf file.
    mts.write_vcf(vcf_file, individual_names=yris+paps+arcs)
```
